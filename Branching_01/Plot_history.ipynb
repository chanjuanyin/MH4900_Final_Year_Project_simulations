## 2 - Plotting the results of our simulation

<a name='2-1'></a>
### 2.1 - Load the relevant csv files

```python
np_dictionary = {}

dir_name = "output"
os_listdir = os.listdir(dir_name)
for file_name in os_listdir:
    file_path = os.path.join(dir_name, file_name)
    data =  pd.read_csv(file_path, delimiter=',', dtype=None, header=None)
    np_dictionary[file_name[:-4]] = data
```

### 2.2 - Plot the PDE in one dimension: fix t in {0, 1, 2, ..., 10} and varies x within [-5, 5]

```python
impose_ylim = False
y_lim = [0,0.2]

real_values = np.asarray(np_dictionary["t_equals_0"])
real_values = np.reshape(real_values, (901,))
x_values = np.arange(1, 10.01, 0.01)
# estimated_values = np_dictionary["estimated_values"][1:, 0]
# x_estimated = np.arange(1, 11, 1)

plt.figure(figsize=(12, 8))
plt.plot(x_values, real_values, label='Real values of u(0,x)')
# plt.plot(x_estimated, estimated_values, 'ro-', label='Estimated values')

plt.ylabel('real vs estimated values')
if impose_ylim:
    plt.ylim(y_lim)
plt.xlabel('x values')
# plt.axvline(x=30,color='red')

plt.title('Real vs estimated values of u(0,x)')
plt.legend(loc='lower right')
plt.show()
```
