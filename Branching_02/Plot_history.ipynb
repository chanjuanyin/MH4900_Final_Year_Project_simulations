```python
# From system
import os
import shutil
from pathlib import Path
import random

# Data structure
import numpy as np
import pandas as pd
import json
import csv

# Image processing
import cv2
from matplotlib import pyplot as plt
```

<a name='2'></a>
## 2 - Plotting the results of our simulation

<a name='2-1'></a>
### 2.1 - Load the relevant csv files

```python
np_dictionary = {}

dir_name = "output"
os_listdir = os.listdir(dir_name)
for file_name in os_listdir:
    file_path = os.path.join(dir_name, file_name)
    data =  pd.read_csv(file_path, delimiter=',', dtype=None, header=None)
    np_dictionary[file_name[:-4]] = data
```

<a name='2-3'></a>
### 2.2 - Plot the PDE in one dimension: fix x in {2, 3, ..., 10} and varies t within [0, x-1]

```python
impose_ylim = False
y_lim = [0,0.2]

real_values = np.asarray(np_dictionary["n_equals_3_x_equals_2"])
real_values = np.reshape(real_values, (1001,))[0:101]
t_values = np.arange(0, 1.01, 0.01)
estimated_values = np.asarray(np_dictionary["n_equals_3_x_equals_2_monte_carlo"])
estimated_values = np.reshape(estimated_values, (11,))
t_estimated = np.arange(0, 1.1, 0.1)

plt.figure(figsize=(12, 8))
plt.plot(t_values, real_values, label='Real values of u(t,2)')
plt.plot(t_estimated, estimated_values, 'ro-', label='Estimated values')

# initial_condition_real_values = np.asarray(np_dictionary["x_equals_0"])
# initial_condition_real_values = np.reshape(initial_condition_real_values, (901,))
# t_values_initial = np.arange(1, 10.01, 0.01)
# plt.plot(t_values_initial, initial_condition_real_values, label='Initial condition u(t,0)')

plt.ylabel('real vs estimated values')
if impose_ylim:
    plt.ylim(y_lim)
plt.xlabel('t values')
# plt.axvline(x=30,color='red')

plt.title('Real vs estimated values of u(t,2)')
plt.legend(loc='upper right')
plt.show()
```
