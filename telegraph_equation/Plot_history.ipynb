## 2 - Plotting the results of our simulation

<a name='2-1'></a>
### 2.1 - Load the relevant csv files

```python
np_dictionary = {}

dir_name = "output"
os_listdir = os.listdir(dir_name)
for file_name in os_listdir:
    file_path = os.path.join(dir_name, file_name)
    data = np.genfromtxt(file_path, delimiter=',', dtype=None)
    np_dictionary[file_name[:-4]] = data
```

<a name='2-2'></a>
### 2.2 - Plot the PDE in one dimension: fix t in {0, 1, 2, ..., 10} and varies x within [-5, 5]

```python
impose_ylim = False
y_lim = [0,0.2]

real_values = np_dictionary["t_equals_0"]
x_values = np.arange(-5, 5.01, 0.01)
estimated_values = np_dictionary["estimated_values"][:, 0]
x_estimated = np.arange(-5, 6, 1)

plt.figure(figsize=(12, 8))
plt.plot(x_values, real_values, label='Real values')
plt.plot(x_estimated, estimated_values, 'ro-', label='Estimated values')

plt.ylabel('F(x,0)')
if impose_ylim:
    plt.ylim(y_lim)
plt.xlabel('x values')
# plt.axvline(x=30,color='red')

plt.title('Real vs estimated values of F(x,0)')
plt.legend(loc='lower right')
plt.show()
```
